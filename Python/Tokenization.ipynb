```python
import nltk
from nltk.tokenize import word_tokenize
import json

def tokenizer_nltk_fichier_text(fichier, langue):
    liste_tokens = []
    with open(fichier, "r", encoding="utf-8") as fin:
        contenu = fin.read()
        liste_tokens = word_tokenize(contenu, language=langue)
    return(liste_tokens)

def tokenizer_nltk_text(texte, langue):
    liste_tokens = word_tokenize(texte, language=langue)
    return(liste_tokens)

def lire_fichier(fichier):
    with open(fichier, "r", encoding="utf-8") as fin:
        contenu = fin.read()
    return contenu

def ecrire_json(chemin , contenu):
    with open(chemin , "w", encoding="utf-8") as fout:
        fout.write(json.dumps(contenu , indent=2, ensure_ascii=False))
        
def lire_json(chemin):
    dic = {}
    with open(chemin , "r", encoding="utf-8") as fin:
        dic = json.load(fin)
    return dic
```

```python
chemin_correction = "Evaluation/Correction/Correction/PERRROT75CORR.txt"

chemin_base = "Evaluation/Correction/A_corriger/PERRROT75.txt"

tokens_base = tokenizer_nltk_fichier_text(chemin_base, "french")
print(len(tokens_base))

tokens_correction = tokenizer_nltk_fichier_text(chemin_correction, "french")
print(len(tokens_correction))
```

```python
from bs4 import BeautifulSoup

def tokens_from_xml(fichier):
    liste_tokens = []
    with open(fichier, "r", encoding="utf-8") as fin:
        contenu = fin.read()
        soup = BeautifulSoup(contenu, 'html.parser')
        for token in soup.find_all("token"):
            #if token["kind"] == "word":
            liste_tokens.append(token.text)
    return liste_tokens
```

```python
fichier = "Evaluation/Correction/Tokenization/PERRROT75.xml"

fichier_correction = "Evaluation/Correction/Tokenization/PERRROT75CORR.xml"

tokens_base = tokens_from_xml(fichier)

tokens_correction = tokens_from_xml(fichier_correction)

print(len(tokens_base))

print(len(tokens_correction))
```
